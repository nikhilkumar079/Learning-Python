```python
import bokeh.io
import bokeh.plotting
bokeh.io.output_notebook()
```

Q1. How can you create a Bokeh plot using Python code?

```python
from bokeh.plotting import figure, output_file, show
x = [1, 2, 3, 4, 5]
y = [6, 7, 2, 4, 5]

p = figure(title='Example Bokeh Plot', x_axis_label='X Axis', y_axis_label='Y Axis')
p.line(x, y, legend_label='Line', line_width=2)

output_file('example.html')

show(p)
```

Q2. What are glyphs in Bokeh, and how can you add them to a Bokeh plot ?
 Explain with an example.

In Bokeh, glyphs are the basic visual building blocks used to create plots. Glyphs are graphical representations of data, such as points, lines, and bars. Bokeh provides a wide range of glyphs that can be used to create different types of plots.

```python
from bokeh.models.annotations import Legend
from bokeh.models import ColumnDataSource
x = [1, 2, 3, 4, 5]
y = [6, 7, 2, 4, 5]
colors = ['red', 'green', 'blue', 'yellow', 'black']

source = ColumnDataSource(data=dict(x=x, y=y, color=colors))

p = figure(title='Plot using Glyphs', x_axis_label='X Axis', y_axis_label='Y Axis')

p.circle('x', 'y', size=15, fill_color='color', line_color='black', source=source, legend_label='point')

output_file('glyphs.html')

show(p)
```

Q3. How can you customize the appearance of a Bokeh plot, including the axes, title, and legend?

```python
# Customizing the appearance of the axes
p.xaxis.axis_label_text_font_size = '16pt'
p.yaxis.axis_label_text_font_size = '16pt'
p.xaxis.major_tick_line_color = 'blue'
p.yaxis.major_tick_line_color = 'blue'
p.xaxis.minor_tick_line_color = 'red'
p.yaxis.minor_tick_line_color = 'red'

# Customizing the appearance of the title
p.title.text_font_size = '18pt'
p.title.text_color = 'Blue'

# Customizing the appearance of the legend
p.legend.location = 'top_left'
p.legend.label_text_font_size = '14pt'
p.legend.background_fill_color = 'lightgray'

show(p)
```

Q4. What is a Bokeh server, and how can you use it to create interactive plots that can be updated in real time?

Bokeh server is a Python library that allows you to create interactive web-based visualizations that can be updated in real-time. With Bokeh server, you can create rich and complex applications that allow users to interact with data in a flexible and intuitive way.

To use Bokeh server to create interactive plots, you need to first define a Bokeh figure and add the necessary glyphs and tools to it, just like you would with a regular Bokeh plot. However, instead of calling show() to display the plot, you will create a Bokeh server application that serves the plot to a web page, where it can be accessed and interacted with by users.
