Q1. Explain GET and POST methods.

GET and POST are HTTP methods that are used to submit data from a client (such as a web browser) to a server (such as 

a web application).

The GET method is used to request a resource or information from the server. It sends data through the URL in the form 
of query parameters, which are visible in the address bar of the browser. 

The POST method is used to submit data to be processed by the server. It sends data in the request body, which is not 

visible in the address bar of the browser. POST requests are typically used to submit data, such as a form, to be 

processed by the server. Since the data is sent in the request body, POST requests have a higher capacity to send 

larger amounts of data. POST requests are also considered to be more secure as the data is not visible in the URL.

Q2. Why is request used in Flask?

In Flask, the request object is used to access incoming data from a client's HTTP request. It provides access to the 

headers, URL, form data, and other parameters of the request.

Q3. Why is redirect() used in Flask?

In Flask, the redirect() function is used to redirect a client's request to a different URL. It sends a response to 

the client with a new URL, and the client's browser will automatically make a new request to the new URL.

from flask import Flask, redirect, url_for

#### EXAMPLE

app = Flask(__name__)

@app.route('/')

def index():

    return redirect(url_for('hello'))

@app.route('/hello')

def hello():

    return 'Hello, World!'


In the above example, when a client visits the root URL (/), they will be redirected to the hello() function using 

redirect() and url_for().

Q4. What are templates in Flask? Why is the render_ template) function used?

In Flask, templates are used to generate dynamic HTML pages that can display data from a web application. A template 
is essentially an HTML file that includes placeholders for dynamic content, which can be filled in with data from a 
Flask application.

In Flask, the render_template() function is used to render a Jinja2 template with the specified context data. It takes 
the name of the template file and any data that should be passed to the template as arguments.
