Q2: What is a scatter plot? Use the given code to generate data for x and y. Using this generated data plot a scatter plot.

Scatter plot is used to display the relationship between two variables by plotting individual data points on a two-dimensional plane.

```python
#Given Code
np.random.seed(3)
x = 3 + np.random. normal(0, 2, 50)
y =3 + np.random.normal(0, 2, len(x))
```

```python
plt.scatter(x, y)

plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Scatter plot of data x and y')

plt.show()
```

Q3: Why is the subplot function used? Draw four line plots using the subplot() function.

The subplot() function in Matplotlib is used to create multiple plots within the same figure. It allows us to display different visualizations side by side or in a grid, making it easier to compare and analyze multiple datasets or variables.

```python
#Given Data
x1 = np.array ([0, 1, 2, 3, 4, 5])
y1 = np.array([0, 100, 200, 300, 400, 500])
x2 = np.array ([0, 1, 2, 3, 4, 5])
y2 = np.array ([50, 20, 40, 20, 60, 70])
x3 = np.array([0, 1, 2, 3, 4, 5])
y3 = np.array([10, 20, 30, 40, 50, 60])
x4 = np.array ([0, 1, 2, 3, 4, 5])
y4 = np.array ([200, 350, 250, 550, 450, 150])
```

```python
fig, ax = plt.subplots()
ax.plot(x1, y1, label='Line 1')
ax.plot(x2, y2, label='Line 2')
ax.plot(x3, y3, label='Line 3')
ax.plot(x4, y4, label='Line 4')
ax.set_xlabel('X-axis')
ax.set_ylabel('Y-axis')
ax.set_title('Four lines plot')
ax.legend()
plt.show()
```

```python
#2x2 grid of subplots
plt.subplot(2, 2, 1)
plt.plot(x1, y1)
plt.title('Line 1')

plt.subplot(2, 2, 2)
plt.plot(x2, y2)
plt.title('Line 2')

plt.subplot(2, 2, 3)
plt.plot(x3, y3)
plt.title('Line 3')

plt.subplot(2, 2, 4)
plt.plot(x4, y4)
plt.title('Line 4')

plt.tight_layout()
plt.show()
```

Q4: What is a bar plot? Why is it used? Using the given data plot a bar plot and a horizontal bar plot.

A bar plot is a graphical representation of data that uses rectangular bars to compare the sizes of data. It is a common plot type used to display categorical data.

```python
#Given
company = np.array (["Apple", "Microsoft", "Google", "AMD"])
profit = np.array ([3000, 8000, 1000, 10000])
```

```python
fig, ax = plt.subplots()
ax.bar(company, profit)
ax.set_xlabel('Company')
ax.set_ylabel('Profit')
ax.set_title('Companies Profit')
plt.show()
```

```python
fig, ax = plt.subplots()
ax.barh(company, profit)
ax.set_xlabel('Company')
ax.set_ylabel('Profit')
ax.set_title('Companies Profit')
plt.show()
```

Q5: What is a box plot? Why is it used? Using the given data plot a box plot.

A box plot is a graphical representation of the distribution of a dataset. It displays the median, quartiles, and outliers of the data in a concise way.

```python
#Given
box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)
```

```python
fig, ax = plt.subplots()

ax.boxplot([box1,box2])
ax.set_xticklabels(['Box 1', 'Box 2'])
ax.set_ylabel('Values')
ax.set_title('Box Plot')
plt.show()
```
