Q4. Explain with an example:

    a. try and else
    
    b. finally
    
    c. raise

```python
#try and else
# else keyword in a try - except block to define what to do if no errors were raised
def number(x):
    try:
        if x > 1000:
            raise ValueError(x)
    except ValueError:
        print(x, "is out of allowed range. Please enter number below 1000")
    else:
        print(x, "is within the allowed range")
```

```python
number(100)
```

```python
number(10000)
```

```python
#finally
#The finally block will always be executed, no matter if the try block raises an error or not.
def divide(x, y):
        try:
            result = x / y
        except ZeroDivisionError:
            print("Error: Division by zero doesn't make any sense!")
        else:
            print("result is", result)
        finally:
            print("The task is completed")
```

```python
divide(2,1)
```

```python
divide(10,0)
```

Q5. What are Custom Exceptions in python? Why do we need Custom Exceptions? Explain with an example.

Custom exception are user-defined for a particular purpose. They may not be built in errors, but they are useful when we are working on specific data. Also, custom exceptions provide you the flexibility to add attributes and methods that are not part of built-in exceptions.

```python
#Example
try:
    age = int(input())
except:
    print("Invalid Age")
else:
    print(age) #the age is printed irrespective of value. Because there is no error. Therfore, we need custom exception
```

Q6. Create a custom exception class. Use this class to handle an exception.

```python
class InputError(Exception):
     pass
try:
    x = int(input())
    m =0
    if x < 0:
        raise InputError
    for i in range(1,x+1):
        m+=1
        for j in range (1,i+1):
            print(m, end = " ")
        print("")
except InputError:
    print('Value should be greater than Zero and less than 6')
```
