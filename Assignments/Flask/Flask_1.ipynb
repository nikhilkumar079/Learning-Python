Q1. What is Flask Framework? What are the advantages of Flask Framework?

Flask is a lightweight and flexible web application framework written in Python. It is designed to make it easy to build 

web applications quickly and with minimum hassle.

Some of the advantages of Flask include:

1. Lightweight and flexible: Flask is a minimalistic framework that does not impose any particular way of doing things.
2. Easy to get started: Flask is easy to install and get started with. It has a small codebase and a simple API, which makes it easy to learn.
3. Modular design: Flask is designed in a modular way, with each feature implemented as an extension. This makes it easy to add new functionality to your application.
4. Powerful development server: Flask comes with a powerful built-in development server that makes it easy to develop and test your application.
5. Large ecosystem: Flask has a large and active community of developers who have contributed a vast number of extensions and plugins to the framework.

Q3. What is App routing in Flask? Why do we use app routes?

App routing in Flask refers to the process of mapping a URL to a view function in your Flask application. This 

is typically done using the @app.route decorator, which is used to define a URL route that the application will 
respond to.

We use app routes in Flask to map URLs to specific view functions, so that when a user requests a certain URL, 

the application knows which function to call to generate the appropriate response. App routes allow us to 

create a clean, intuitive URL structure for our application, which makes it easier for users to navigate and 

understand.

Q5. What function is used in Flask for URL Building? Write a Python code to demonstrate the working of the url_for() 
function.

In Flask, the url_for() function is used for URL building. This function generates a URL for the specified endpoint, using 

the endpoint name and any arguments provided.

##### Python Code Demonstrating url_for()
from flask import Flask, url_for

app = Flask(__name__)

@app.route('/hello/<name>')
def hello(name):
    return 'Hello, {}!'.format(name)

with app.test_request_context():
    url = url_for('hello', name='John')
    print(url)
